```python
import pandas as pd
```

```python
train_set = pd.read_excel("/kaggle/input/summerization/Competition CERIST -- Summarization/Dataset 1 Arabic (EACS)/dataset_EASC_v1.0_train.xlsx")
test_set = pd.read_excel("/kaggle/input/summerization/Competition CERIST -- Summarization/Dataset 1 Arabic (EACS)/dataset_EASC_v1.0_test.xlsx")
```

```python
def split_column(df,column, split_car):
# converting to dict
    index_value = df[column].to_dict()
    for key, value in index_value.items():
        if split_car == ",":
            list_index = index_value[key].split(split_car)
            index_value[key] = list(map(int,list_index))
        else :
            list_text = index_value[key].split(split_car)
            index_value[key] = list_text
    return index_value
```

```python
def extract_summary(df):
    summary_text={}
    summary_index = split_column(df, 'Summary', ',')
    document_index = split_column(df, 'Document Sents', '\n')
    for (index1,indices),(index2, texts) in zip(summary_index.items(), document_index.items()):
        summary_text[index1] = ""
        for i in indices :
            summary_text[index1] += texts[i]
    return summary_text
```

```python
train_summary_dict = extract_summary(train_set)
test_summary_dict = extract_summary(test_set)
```

```python
summary_train_df = pd.DataFrame.from_dict(train_summary_dict,orient ='index', columns = ['summary_text'])
summary_test_df = pd.DataFrame.from_dict(test_summary_dict,orient ='index', columns = ['summary_text'])
```

```python
train_df = pd.concat([train_set, summary_train_df], axis=1)
test_df = pd.concat([test_set, summary_test_df], axis=1)
```
